```python
import scapy.all as scapy

def is_abnormal_tcp(tcp_stream):
  # 检查RST标志
  if tcp_stream.has_flags('R'):
    return True

  # 检查三次握手  
  if not tcp_stream.has_flags('S'):
    return True 
  if not tcp_stream.has_flags('SA'):
    return True

  # 检查是否有数据传输
  if len(tcp_stream) < 4:
    return True

  # 检查是否正常关闭
  if not tcp_stream.has_flags('F'):
    return True

  return False

packets = rdpcap('./skype_audio1a.pcap.TCP_131-202-240-150_36188_209-148-205-19_80.pcap')

tcp_packets = packets.filter_packets('tcp')

streams = tcp_packets.sessions() 

abnormal_flows = []

for stream in streams:
  if is_abnormal_tcp(stream):
    abnormal_flows.append(stream)

normal_pkts = [pkt for pkt in tcp_packets if pkt not in abnormal_flows]

scapy.wrpcap('normal_tcp.pcap', normal_pkts)

print(f'{len(abnormal_flows)} abnormal TCP flows filtered')
```
